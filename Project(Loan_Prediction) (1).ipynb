```python
Loandata=loandata.drop(columns=['Loan_ID',	'Gender', 'Married', 	'Education',	'Self_Employed', 'Property_Area','Dependents' ], axis=1)
```

```python
Loandata
```

```python
Loandata.isnull().sum()
```

```python
Loandata['LoanAmount']=Loandata['LoanAmount'].fillna(Loandata['LoanAmount'].mean())
Loandata['Loan_Amount_Term']=Loandata['Loan_Amount_Term'].fillna(Loandata['Loan_Amount_Term'].mean())
Loandata['Credit_History']=Loandata['Credit_History'].fillna(Loandata['Credit_History'].mean())
```

```python
x=Loandata.iloc[:, :-1]
y=Loandata.iloc[:, -1]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y
```

```python
# split data
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1, test_size=0.2)
```

```python
print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(y_test))
```

```python
# Define the model: Init K-NN
classifier = svm.SVC(kernel='linear')
```

```python
# Fit Model
classifier.fit(x_train, y_train)
```

```python
# Predict the test set results
y_pred = classifier.predict(x_test)
y_pred
```

```python
print(accuracy_score(y_test, y_pred))
```
