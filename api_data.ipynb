```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# EDA on Score KPI
sns.histplot(df['score'], kde=True)
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Distribution of Score KPI')
plt.show()

# EDA on Sector KPI
sector_counts = df['sector'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=sector_counts.index, y=sector_counts.values)
plt.xlabel('Sector')
plt.ylabel('Count')
plt.title('Distribution of Sectors')
plt.xticks(rotation=45)
plt.show()

# EDA on Industry KPI
industry_counts = df['industry'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=industry_counts.index, y=industry_counts.values)
plt.xlabel('Industry')
plt.ylabel('Count')
plt.title('Distribution of Industries')
plt.xticks(rotation=45)
plt.show()

# EDA on Exchange KPI
exchange_counts = df['exchange'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=exchange_counts.index, y=exchange_counts.values)
plt.xlabel('Exchange')
plt.ylabel('Count')
plt.title('Distribution of Exchanges')
plt.xticks(rotation=45)
plt.show()
```
